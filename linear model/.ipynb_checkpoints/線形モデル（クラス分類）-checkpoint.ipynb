```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

# ロジスティック回帰、線形サポートベクタマシン

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X,y = mglearn.datasets.make_forge()

fig,axes = plt.subplots(1,2,figsize=(10,3))

for model,ax in zip([LinearSVC(),LogisticRegression()],axes):
    clf = model.fit(X,y)
    mglearn.plots.plot_2d_separator(clf,X,fill=False,eps=0.5,ax=ax,alpha=.7)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title("{}".format(clf.__class__.__name__))
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
axes[0].legend()
```

デフォルトではLinearSVC(線形ベクタマシン)とLogisticRegression(ロジスティック回帰）はRidgeと同様にL2正則化を行う<br>
その正則化の強度を決めるパラメタはCとなっており、<b>C</b>で決定される<br>

```python
mglearn.plots.plot_linear_svc_regularization()
```

線形ベクタマシンのパラメタCを変化させた場合が上図<br>
<b>Cが小さくなるほど正則化が強い。</b>Cを小さくすると水平に近くなっている<br>
一番右だと正則化が弱い。すべての点を正しく分類しようとするあまり傾きが急になっている。おそらく過剰適合していると予想できる。

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)
logreg = LogisticRegression().fit(X_train,y_train)
print("訓練セット:{:.3f}".format(logreg.score(X_train,y_train)))
print("検証セット:{:.3f}".format(logreg.score(X_test,y_test)))
```

上記はロジスティック回帰のパラメタCをデフォルトの1としており、良い精度となっている<br>
しかし訓練セット、検証セットの精度が非常に近いため適合不足の可能性がある。<br>

```python
logreg100 = LogisticRegression(C=100).fit(X_train,y_train)
print("C=100 訓練セット:{:.3f}".format(logreg100.score(X_train,y_train)))
print("C=100 検証セット:{:.3f}".format(logreg100.score(X_test,y_test)))
logreg001 = LogisticRegression(C=0.01).fit(X_train,y_train)
print("C=0.01 訓練セット:{:.3f}".format(logreg001.score(X_train,y_train)))
print("C=0.01 検証セット:{:.3f}".format(logreg001.score(X_test,y_test)))
```

Cを増やし、正則化を弱くすると精度があがる<br>
Cを小さくし、正則化を強くするとデフォルトよりもさらに適合不足となってしまう。

```python
plt.plot(logreg001.coef_.T,'o',label="C=0.01")
plt.plot(logreg.coef_.T,'^',label="C=1")
plt.plot(logreg100.coef_.T,'v',label="C=100")
plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
plt.hlines(0,0,cancer.data.shape[1])

plt.ylim(-5,5)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")
plt.legend()
```

LogisticRegressionではデフォルトでL2正則化が行われる。上記結果はRidgeに似ている。<br>
正則化を強くすると係数は0近くとなるが0にはならない。<br>
上記の3つめの特徴量mean perimeterの係数だがC=100,C=1では係数は負となっているがC=0.01のとき正になっている。<br>
<b>このようなモデルを解消する際には、どの係数がクラス分類に影響を与えているかを考察する</b><br>
例えば特徴量texture errorが大きいことがmalignant（悪性）に関係あるのではないか？と考える<br>
しかしmean perimeterのように正負が変わると解釈の意味も全く変わってしまうので係数の解釈は慎重を要する<br>

解釈しやすいモデルが欲しい場合はL1正則化を使うと良い。<br>
ロジスティック回帰はL2正則化が行われるがL1正則化も行うことができる

```python
for C,marker in zip([0.001,1,100],['o','^','v']):
    lr_l1 = LogisticRegression(C=C,penalty="l1").fit(X_train,y_train)
    print("訓練データ精度 {:.3f}\t{:.2f}".format(C,lr_l1.score(X_train,y_train)))
    print("検証データ精度 {:.3f}\t{:.2f}".format(C,lr_l1.score(X_test,y_test)))
    plt.plot(lr_l1.coef_.T,marker,label="C={:.3f}".format(C))
    
plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
plt.hlines(0,0,cancer.data.shape[1])
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")

plt.ylim(-5,5)
plt.legend(loc=3)
```

上図から2クラス分類の線形モデル郡と、回帰のための線形モデル郡は対応関係がある<br>
回帰の際と同様にモデル間の主な違いはpenaltyパラメタでL1正則化を行うかに違いがある<br>

# 線形モデルによる多クラス分類

多くの線形モデルは2クラス分類しか適用できず、他クラスの場合に拡張できることはない（ロジスティック回帰は除く)<br>
1対多クラスのアプローチでは<b>各クラスに対してそのクラスと他のすべてのクラスを分類する2クラス分類モデルを学習する<br>
クラスがたくさんある場合にはたくさんの2クラス分類モデルを使う</b><br>
つまり2クラス分類を大量に行うことで1対多を実現する

多クラスロジスティック回帰の背後にある数学は1対多へのアプローチとは異なるが結局同じ方法で予測できる<br>

```python
from sklearn.datasets import make_blobs

X,y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0","Class 1","Class 2"])
```

上記データをLinearSVCで分類してみる

```python
linear_svm = LinearSVC().fit(X,y)
print("Coefficient shape:",linear_svm.coef_.shape)
print("intercept shape",linear_svm.intercept_.shape)
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line = np.linspace(-15,15)
for coef,intercept, color in zip(linear_svm.coef_,linear_svm.intercept_,['b','r','g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1],c=color)
plt.ylim(-10,15)
plt.xlim(-10,8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(['Class 0','Class 1','Class 2','Line class 0','Line class 1','Line class 2'],loc=(1.01,0.3))
```

上記図のようにクラス分類されるわけではない<br>
なぜなら、このままでは各クラスを分類するためのラインで作られた中央の三角形の内部では「その他」として分類されてしまう<br>
どのクラスに分類すればよいか？それは「クラス分類式の値が一番大きいクラス」つまり、その点に最も近い線を持つクラスとなる<br>
そのため、最終的な境界線は以下となる。

```python
mglearn.plots.plot_2d_classification(linear_svm,X,fill=True,alpha=.7)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line = np.linspace(-15,15)

for coef,intercept,color in zip(linear_svm.coef_,linear_svm.intercept_,['b','r','g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1],c=color)
plt.legend(['Class 0','Class 1','Class 2','Line class 0','Line class 1','Line class 2'],loc=(1.01,0.3))
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

# まとめ

線形モデルの主要なパラメタ
線形、リッジ、ロッソ→alpha<br>
線形SVC、ロジスティック回帰→C<br>
これらのパラメタの調整が重要となる。通常Cやalphaを調整する場合、対数スケールで値を変更する。<br>
また、L1正則化を使うか、L2正則化を使うかを決めることが重要。<br>
一部の特徴量だけが重要だと考えるならL1を使用するが、それ以外の場合はL2を使用したほうが良い。<br><br>
線形モデルの訓練は非常に高速で予測も早い。非常に大きいデータセットにも適用できるし、疎なデータに対してもうまく機能する<br>
<b>ただしサンプルが10万、100万などある場合はsolver = 'sag'オプションの使用も検討すること。これを使用すると大きなデータセットに対して高速に処理できることがある</b><br>
係数がどうしてその値となるかについてはそれほど明らかでない。<br>
線形モデルは特徴量の数がサンプルの個数よりも多いときに性能を発揮する。大きなデータセットに用いられることも多いが単に他のモデルでは学習できないからである<br>
しかし低次元空間では他のモデルのほうが良い汎化性能を占めることもある。（サポートベクタマシンなど）
